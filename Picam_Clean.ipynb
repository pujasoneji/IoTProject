```python
docs = pandas.DataFrame(columns=[])
for num, doc in enumerate(mongo_docs):
    doc["_id"] = str(doc["_id"])
    doc_id = doc["_id"]
    
    series_obj = pandas.Series( doc, name=doc_id )
    docs = docs.append(series_obj)

print(docs)
```

```python
start = (docs.iloc[1]['time'])-1 #gives the time for the first entry
#print(start)
#end = start + 90*len(mongo_docs) + 90
#end = docs.iloc[-1]['time'] + 90
bins = np.arange(start, 1610496259, 90)
category = []
for n in bins:
    string = str(n)[:-2] #gets rid of .0
    category.append(string)
del category[-1] #there needs to be one less category than no. of bins
#print(category)
```

```python
col_picam = db["collection_picam"] #name of collection on mongodb
col_all = db["collection_all"]
cursor_picam = col_picam.find()
mongo_docs_picam = list(cursor_picam)
count = 4 #starts from viable data
x = 0
q = True
loop = False

while q:
    if count < len(mongo_docs_picam):
        event = mongo_docs_picam[count]
        detected = int(event['detected'])
        if loop is False:
            if int(event['time']) > int(category[x]) and int(event['time']) < int(category[x+1]): #event starts here
                col_all.update_one({"_id":str(category[x])},{"$set":{"detected":detected}},upsert=False)
                #print("Detection starts", detected)
                print(detected)
                loop = True
        if loop is True:
            futureevent= mongo_docs_picam[count+1]
            #print(futureevent, category[x+1])
            if (int(futureevent['time'])) > (int(category[x+1])):
                col_all.update_one({"_id":str(category[x])},{"$set":{"detected":detected}},upsert=False)
                #print("Still detecting", detected)
                print(detected)
            elif int(futureevent['time']) < int(category[x+1]):
                col_all.update_one({"_id":str(category[x])},{"$set":{"detected":detected}},upsert=False)
                #print("Last detection", detected)
                print(detected)
                count += 1
                loop is False
    else:
        if x < len(category)-1:
            col_all.update_one({"_id":str(category[x])},{"$set":{"detected":0}},upsert=False) #records that I'm not sitting
            print("no more data")
        else:
            q = False
    x += 1
print("Done")
    
```
